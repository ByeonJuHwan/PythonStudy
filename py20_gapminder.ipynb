* (중복되지 않는) 국가 이름의 개수?
* (중복되지 않는) 대륙 이름?
* (중복되지 않는) 연도의 개수?
* 국가이름, 연도, 1인당 GDP 컬럼을 선택
* 1인당 GDP의 최댓값, 최솟값을 찾고, 그 row를 출력
* 기대수명(lifeExp)의 최솟값과 최댓값을 찾고, 그 행들을 출력
* 연도별 기대수명의 평균을 출력
* 연도별 1인당 GDP의 평균을 출력
* 연도별 대륙별 전세계 인구를 출력


```python
import pandas as pd
import seaborn as sns
```

```python
file_path = 'https://github.com/JakeOh/202208_itw_java134_lab_python/raw/main/gapminder.tsv'
con = pd.read_csv(file_path, sep='\t')
con
```

```python
# (중복되지 않는) 국가의 개수
res = 0
for _ in con.country.unique():
    res+=1
res
```

```python
# (중복되지 않는) 대륙의 개수
res = 0
for _ in con.continent.unique():
    res+=1
res
```

```python
# (증복되지 않는) 연도의 개수
res = 0
for _ in con.year.unique():
    res+=1
res
```

```python
# 국가이름, 연도, 1인당 GDP컬럼 선택
con[['country','year','gdpPercap']]
```

```python
# 1인당 GDP의 최댓값, 최솟값의 row
con[(con.gdpPercap == con.gdpPercap.max())|
    (con.gdpPercap == con.gdpPercap.min())
    ]
```

```python
# 기대수명(lifeExp)의 최솟값과 최댓값을 찾고, 그 행들을 출력
con[(con.lifeExp == con.lifeExp.max())|(con.lifeExp == con.lifeExp.min())]
```

```python
# 연도별 기대수명의 평균을 출력
con.groupby('year')[['lifeExp']].mean()
```

```python
# 연도별 1인당 GDP의 평균을 출력
con.groupby('year')[['gdpPercap']].mean()
```

```python
# 연도별 대륙별 전세계 인구를 출력
con.groupby(['year','continent'])[['pop']].sum()
```
